```python
import json
def load_data(filename):
    with open("data.json","r") as f:
        return json.load(f)
```

```python
def people_you_may_know(user_id,data ):
    user_friends={}
    for user in data['users']:
        user_friends[user['id']]=set(user['friends'])
    if user_id not in user_friends:
        return

    dict_friends=user_friends[user_id]
    suggestion={}
    for friend in dict_friends:
        for mutual in user_friends[friend]:
            if mutual!=user_id and mutual not in dict_friends:
                suggestion[mutual]=suggestion.get(mutual,0)+1 
    sorted_suggestion=sorted(suggestion.items(),key=lambda x:x[1],reverse=True)
    return[user_id for user_id, mutual_count in sorted_suggestion]
    
     
#load the data
data =load_data("data.json")
user_id=1
record=people_you_may_know(user_id,data)
print(record)
```
